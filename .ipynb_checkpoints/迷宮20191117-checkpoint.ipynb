### 老鼠走迷宮

```python
### 建立題目串列
maze = []
maze.append([1,1,1,1,1,1])
maze.append([1,0,1,1,1,1])
maze.append([1,1,0,1,1,1])
maze.append([1,0,1,0,0,1])
maze.append([1,1,0,1,0,1])
maze.append([1,1,1,1,1,1])

### 建立判別串列
judge = []
judge.append([1,1,1,1,1,1])
judge.append([1,1,0,0,0,1])
judge.append([1,0,0,0,0,1])
judge.append([1,0,0,0,0,1])
judge.append([1,0,0,0,0,1])
judge.append([1,1,1,1,1,1])

### 建立行走堆疊
movestack = []
movestack.append([1,1,1])

### 移動代號的對應(右上->1,右->2,右下->3,D下->4,左下->5,左->6,左上->7,上->8)

### 堆疊檢驗
while(True):
    if (len(movestack)==0):
        break
    if (movestack[-1][0]==4 & movestack[-1][1]==4):
        break
    h = movestack[-1][0]
    v = movestack[-1][1]
    if (movestack[-1][2] == 1):###右上
        if ((judge[h-1][v+1] ==1) or (maze[h-1][v+1] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h-1][movestack[-1][1]+1] =1
            movestack.append([h-1,v+1,1])
    elif (movestack[-1][2] == 2):###右
        if ((judge[h][v+1] ==1) or (maze[h][v+1] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h][v+1] =1
            movestack.append([h,v+1,1])
    elif (movestack[-1][2] == 3):###右下
        if ((judge[h+1][v+1] ==1) or (maze[h+1][v+1] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h+1][v+1] =1
            movestack.append([h+1,v+1,1])
    elif (movestack[-1][2] == 4):###下
        if ((judge[h+1][v] ==1) or (maze[h+1][v] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h+1][v] =1
            movestack.append([h+1,v,1])           
    elif (movestack[-1][2] == 5):###左下
        if ((judge[h+1][v-1] ==1) or (maze[h+1][v-1] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h+1][v-1] =1
            movestack.append([h+1,v-1,1])
    elif (movestack[-1][2] == 6):###左
        if ((judge[h][v-1] ==1) or (maze[h][v-1] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h][v-1] =1
            movestack.append([h,v-1,1])
    elif (movestack[-1][2] == 7):###左上
        if ((judge[h-1][v-1] ==1) or (maze[h-1][v-1] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h-1][v-1] =1
            movestack.append([h-1,v-1,1])         
    elif (movestack[-1][2] == 8):###上
        if ((judge[h-1][v] ==1) or (maze[h-1][v] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[[h]-1][v] =1
            movestack.append([h-1,v,1])            
    else:
        movestack.pop(-1)        
        
### 輸出結果       
if len(movestack) == 0:
    print('No solution!!')
else:
    ans = []
    for i in range (6):
        ans.append([1,1,1,1,1,1])
    for i in movestack:
        ans[i[0]][i[1]]=0
    print('迷宮')
    for i in maze:
        print(i)
    print('解答路徑')
    for i in ans:
        print(i)
    
```

### 生成+破解

```python
### 生成迷宮
import random
n = int(input('請輸入n值來產生隨機的n x n迷宮：'))
n = n + 2
maze = []
top = [1]*n
maze.append(top)
for i in range (n-2):
    temp = []
    temp.append(1)
    for j in range (n-2):
        if (random.randint(1,100)%3 == 0):
            temp.append(1)
        else:
            temp.append(0)
    temp.append(1)
    maze.append(temp)
maze.append(top)
if maze[1][1]==1:
    maze[1][1]=0
if maze[n-2][n-2]==1:
    maze[n-2][n-2]=0

### 破解迷宮

### 建立判別串列
judge = []
mid = []
for i in range(n-3):
    mid.append(0)

judge.append(top)
judge.append([1]+[1]+mid+[1])
for i in range(n-3):
    judge.append([1]+[0]+mid+[1])
judge.append(top)

### 建立行走堆疊
movestack = []
movestack.append([1,1,1])

### 移動代號的對應(右上->1,右->2,右下->3,D下->4,左下->5,左->6,左上->7,上->8)

### 堆疊檢驗
while(True):
    if (len(movestack)==0):
        break
    if (movestack[-1][0]==n-2 & movestack[-1][1]==n-2):
        break
    h = movestack[-1][0]
    v = movestack[-1][1]
    if (movestack[-1][2] == 5):###右上
        if ((judge[h-1][v+1] ==1) or (maze[h-1][v+1] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h-1][movestack[-1][1]+1] =1
            movestack.append([h-1,v+1,1])
    elif (movestack[-1][2] == 3):###右
        if ((judge[h][v+1] ==1) or (maze[h][v+1] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h][v+1] =1
            movestack.append([h,v+1,1])
    elif (movestack[-1][2] == 1):###右下
        if ((judge[h+1][v+1] ==1) or (maze[h+1][v+1] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h+1][v+1] =1
            movestack.append([h+1,v+1,1])
    elif (movestack[-1][2] == 2):###下
        if ((judge[h+1][v] ==1) or (maze[h+1][v] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h+1][v] =1
            movestack.append([h+1,v,1])           
    elif (movestack[-1][2] == 4):###左下
        if ((judge[h+1][v-1] ==1) or (maze[h+1][v-1] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h+1][v-1] =1
            movestack.append([h+1,v-1,1])
    elif (movestack[-1][2] == 6):###左
        if ((judge[h][v-1] ==1) or (maze[h][v-1] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h][v-1] =1
            movestack.append([h,v-1,1])
    elif (movestack[-1][2] == 8):###左上
        if ((judge[h-1][v-1] ==1) or (maze[h-1][v-1] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h-1][v-1] =1
            movestack.append([h-1,v-1,1])         
    elif (movestack[-1][2] == 7):###上
        if ((judge[h-1][v] ==1) or (maze[h-1][v] ==1)):
            movestack[-1][2] = movestack[-1][2] + 1
        else:
            judge[h-1][v] =1
            movestack.append([h-1,v,1])            
    else:
        movestack.pop(-1)        

### 輸出結果       
if len(movestack) == 0:
    print('No solution!!')
else:
    ans = []
    for i in range (n):
        ans.append([11]*n)
    c = 20
    for i in movestack:
        ans[i[0]][i[1]]=c
        c = c + 1
    print('迷宮')
    for i in maze:
        print(i)
    print('解答路徑')
    for i in ans:
        print(i)
```
